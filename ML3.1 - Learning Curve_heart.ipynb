```python
# Train a decision tree with max_depth=2

## The maximum depth of the tree. If None, then nodes are expanded until all leaves are pure 
## or until all leaves contain less than min_samples_split samples.

clf = tree.DecisionTreeClassifier(max_depth=2)
clf.fit(X, y)
```

Compute the predictions of a decision tree classifier on a test set and calculate all the classification metrics you know:


```python
# split in train and test set
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# fit classifier with max_depth=2

clf.fit(X_train, y_train)
```

```python
# make predictions on the test set
y_pred_clf = clf.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, precision_score, recall_score, accuracy_score, f1_score

# Print Accuracy
print(f"Accuracy metrics: {accuracy_score(y_test, y_pred_clf)}")
# Print Precision
print(f"Precision Score: {precision_score(y_test, y_pred_clf, average='micro')}")
# Print Recall
print(f"Recall: {recall_score(y_test, y_pred_clf, average='micro')}")
# Print F1-score
print(f"F1 Score: {f1_score(y_test, y_pred_clf, average='micro')}")
# Print Confusion matrix
print(f"Confusion matrix:\n {confusion_matrix(y_test, y_pred_clf)}")
```

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams

rcParams['figure.figsize'] = 3,3

plt.bar(df_heart['target'].unique(), df_heart['target'].value_counts(), width=0.4, color = ['red', 'white'])
plt.xticks([0, 1])
plt.xlabel('Target Classes')
plt.ylabel('Count')
plt.title('Count of each Target Class')
```

```python
multicategorical_features = ['cp', 'restecg', 'slope', 'ca', 'thal']
continuous_features = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
```

```python
#getting dummies

df_heart_dummies= pd.get_dummies(df_heart, columns=multicategorical_features, dtype=float)
df_heart_dummies
```

```python
#Replacing column names

df_heart_dummies = df_heart_dummies.rename(columns={"chol":"cholestrol",
                                "fbs": "fasting blood sugar", 
                                "cp_0": "cp_typical angine",
                                "cp_1": "cp_atypical angine",
                                "cp_2": "non-anginal pain",
                                "cp_3": "asymptomatic",
                                "ca_0": "nb_majorvessels_0",
                                "ca_1": "nb_majorvessels_1",
                                "ca_2": "nb_majorvessels_2",
                                "ca_3": "nb_majorvessels_3",
                                "ca_4": "nb_majorvessels_4",
                                "thal_1":"Thalassemia_1",
                                "thal_2":"Thalassemia_2",
                                "thal_3":"Thalassemia_3"})
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler().set_output(transform="pandas")
df_heart_norm = scaler.fit_transform(df_heart_dummies)
print(type(df_heart_norm))
df_heart_norm
```

```python
#Separate features and target 
y = df_heart_dummies['target']
X = df_heart_dummies[['age', 'sex', 'trestbps', 'cholestrol', 'fasting blood sugar',
       'thalach', 'exang', 'oldpeak', 'cp_typical angine',
       'cp_atypical angine', 'non-anginal pain', 'asymptomatic', 'restecg_0',
       'restecg_1', 'restecg_2', 'slope_0', 'slope_1', 'slope_2',
       'nb_majorvessels_0', 'nb_majorvessels_1', 'nb_majorvessels_2',
       'nb_majorvessels_3', 'nb_majorvessels_4', 'Thalassemia_1',
       'Thalassemia_2', 'Thalassemia_3']]
```

```python
#Spliting dataset
X_train_heart, X_test_heart, y_train_heart, y_test_heart = train_test_split(X, y, random_state = 42, 
                                                    test_size = 0.2, 
                                                    stratify = pd.concat([df_heart_norm['sex'], y], axis = 1))
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc=RandomForestClassifier()
rfc.fit(X_train_heart, y_train_heart)
```

```python
y_pred_rfc = rfc.predict(X_test_heart)
```

```python
# Print Accuracy
print(f"Accuracy metrics: {accuracy_score(y_test_heart, y_pred_rfc)}")
# Print Precision
print(f"Precision Score: {precision_score(y_test_heart, y_pred_rfc)}")
# Print Recall
print(f"Recall: {recall_score(y_test_heart, y_pred_rfc)}")
# Print F1-score
print(f"F1 Score: {f1_score(y_test_heart, y_pred_rfc)}")
# Print Confusion matrix
print(f"Confusion matrix:\n {confusion_matrix(y_test_heart, y_pred_rfc)}")
```

## SVM

SVM Advantages:

- SVM’s are very good when we have no idea on the data.
- Works well with even unstructured and semi structured data like text, Images and trees.
- The kernel trick is real strength of SVM. With an appropriate kernel function, we can solve any complex problem.
- Unlike in neural networks, SVM is not solved for local optima.
- It scales relatively well to high dimensional data.
- SVM models have good generalization in practice, the risk of over-fitting is less in SVM.

SVM Disadvantages:

- Choosing a “good” kernel function is not easy.
- Long training time for large datasets.
- Difficult to understand and interpret the final model, variable weights and individual impact.
- The SVM hyper parameters are Cost -C and gamma. It is not that easy to fine-tune these hyper-parameters. It is hard to visualize their impact

```python
from sklearn.model_selection import validation_curve

svm_classifier = SVC()
parameter_range=[1, 10, 20, 30, 40, 50, 60, 70, 80]

train_score, test_score = validation_curve(svm_classifier, X_train_heart,
                                           y_train_heart,
                                           param_range=parameter_range,
                                           param_name="C",
                                           cv=5,
                                           scoring="accuracy")
 
mean_train_score = np.mean(train_score, axis=1)
std_train_score = np.std(train_score, axis=1)
 
mean_test_score = np.mean(test_score, axis=1)
std_test_score = np.std(test_score, axis=1)


fig, ax = plt.subplots(figsize = (8,6))
ax.plot(parameter_range, mean_train_score,
         label="Training Score", color='b')
ax.plot(parameter_range, mean_test_score,
         label="Cross Validation Score", color='g')
ax.legend(loc="upper right")
ax.set_title("Validation Curve with svm_classifier")
ax.set_xlabel("C parameter")
ax.set_ylabel("Accuracy Score")
ax.legend()
ax.grid(True)
plt.show()

```
